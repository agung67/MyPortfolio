```python
def dataconv_indo(dfc):
  lst = list(dfc.columns) 
  lst = lst[4::]
  #lst
  lst2 = list (dfc.loc[dfc.loc[dfc['Country/Region']=='Indonesia'].index[0]])
  lst2 = lst2[4::]
  lst2
  dict = {'Datetime' : lst, 'Case': lst2}
  df = pd.DataFrame(dict)
  df['Datetime'] = pd.to_datetime(df['Datetime'])
  return df

def dataconv_china(dfc):
  lst = list(dfc.columns) 
  lst = lst[4::]
  #lst
  lst2 = list (dfc.loc[dfc.loc[dfc['Country/Region']=='China'].index[0]])
  lst2 = lst2[4::]
  lst2
  dict = {'Datetime' : lst, 'Case': lst2}
  df = pd.DataFrame(dict)
  df['Datetime'] = pd.to_datetime(df['Datetime'])
  return df

def dataconv_aus(dfc):
  lst = list(dfc.columns) 
  lst = lst[4::]
  #lst
  lst2 = list (dfc.loc[dfc.loc[dfc['Country/Region']=='Australia'].index[0]])
  lst2 = lst2[4::]
  lst2
  dict = {'Datetime' : lst, 'Case': lst2}
  df = pd.DataFrame(dict)
  df['Datetime'] = pd.to_datetime(df['Datetime'])
  return df
```

```python

df_indoconf = dataconv_indo(df_confir)
df_indorecov = dataconv_indo(df_recovered)
df_indodeath = dataconv_indo(df_deaths)

df_chinaconf = dataconv_china(df_confir)
df_chinarecov = dataconv_china(df_recovered)
df_chinadeath = dataconv_china(df_deaths)

df_ausconf = dataconv_aus(df_confir)
df_ausrecov = dataconv_aus(df_recovered)
df_ausdeath = dataconv_aus(df_deaths)

df_ausconf
```

```python
import numpy as np

from bokeh.layouts import gridplot
from bokeh.plotting import figure, output_file, show

TOOLS = "pan,wheel_zoom,box_zoom,reset,save,box_select"

p12 = figure(tools=TOOLS)

p12.line(df_indoconf['Datetime'], df_indoconf['Case'],legend_label="Indo Confirmed")
p12.line(df_indodeath['Datetime'], df_indodeath['Case'],legend_label="Indo Death", color="orange")
p12.line(df_indorecov['Datetime'], df_indorecov['Case'],legend_label="Indo Recovered", color = "green")


p12.legend.title = 'Covid-19 in Indonesia'

```

```python

TOOLS = "pan,wheel_zoom,box_zoom,reset,save,box_select"

p22 = figure(tools=TOOLS)

p22.line(df_ausconf['Datetime'], df_ausconf['Case'],legend_label="Australia Confirmed")
p22.line(df_ausdeath['Datetime'], df_ausdeath['Case'],legend_label="Australia Death", color="orange")
p22.line(df_ausrecov['Datetime'], df_ausrecov['Case'],legend_label="Australia Recovered", color = "green")


p22.legend.title = 'Covid-19 in Australia'
```

```python
TOOLS = "pan,wheel_zoom,box_zoom,reset,save,box_select"

p32 = figure(tools=TOOLS)

p32.line(df_chinaconf['Datetime'], df_chinaconf['Case'],legend_label="China Confirmed")
p32.line(df_chinadeath['Datetime'], df_chinadeath['Case'],legend_label="China Death", color="orange")
p32.line(df_chinarecov['Datetime'], df_chinarecov['Case'],legend_label="China Recovered", color = "green")


p32.legend.title = 'Covid-19 in China'
```

```python
tab_indo= Panel(child=p12, title='Covid-19 in Indonesia')
tab_aus = Panel(child=p22, title='Covid-19 in Australia')
tab_china= Panel(child=p32, title="Covid-19 in China")
country = Tabs(tabs=[ tab_indo,tab_aus,tab_china])
```
