```python
import pandas as pd
pd.set_option('display.max_colwidth', None)
```

```python
df1 = pd.read_csv('BLM1.csv')
```

```python
df2 = pd.read_csv('BLM2.csv')
```

```python
df3 = pd.read_csv('BLM3.csv')
```

```python
df1['week'] = '1'
df1['class'] = 0
#origin_index = df1.index.tolist()
#df1 = df1.append(df1, ignore_index=True)
```

```python
df2['week'] = '2'
df2['class'] = 0
```

```python
df3['week'] = '3'
df3['class'] = 0
```

```python
def classification(x):
    for i in range(len(x)):
        if x['retweet_count'][i]==0:
            x.loc[i, ('class')]=0
        elif x['retweet_count'][i]<100:
            x.loc[i, ('class')]=1
        elif x['retweet_count'][i]<1000:
            x.loc[i, ('class')]=2
        elif x['retweet_count'][i]>=1000:
            x.loc[i, ('class')]=3
    return(x)
```

```python
df1=classification(df1)
df2=classification(df2)
df3=classification(df3)
```

```python
test=df3[['retweet_count','class']]
```
