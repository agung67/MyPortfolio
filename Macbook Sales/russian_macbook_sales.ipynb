```python
import pandas as pd
import numpy as np
import glob
```

<span style="font-family:Calibri; font-size:18pt;">Combine dataset into a single dataframe</span>

```python
csvfiles = []
csvfiles = glob.glob("macbooks/*.csv")
print(csvfiles)
```

```python
test = pd.read_csv(csvfiles[0],index_col=0)
test
```

<span style="font-family:Calibri; font-size:18pt;">Get date and month from csv filename</span>

```python
df1 = pd.read_csv(csvfiles[35],index_col=0)
str(csvfiles[34][-10:-4])
```

```python
for i in range(35,len(csvfiles)):
    suf = str(csvfiles[i][-10:-4])
    df2 = pd.read_csv(csvfiles[i],index_col=0)
    df1 = df1.merge(
    
        df2[['name','price']],
        how ='right',
        on = 'name',
        suffixes = ('',suf),
    
    )
    
```

```python
df = df1
```

```python
df
```

```python
df=df.drop("link",axis=1)
```

```python
df
```

<span style="font-family:Calibri; font-size:18pt;">Change "In Stock" and "Out of Stock" into 1 and  respectively</span>

```python
def status(x):
    for i in range(len(x)):
        if x['status'][i]=="In stock":
            x.loc[i, ('status')]=1
        elif x['status'][i]=="Out of stock":
            x.loc[i, ('status')]=0
    return(x)
```

```python
df['status'][2]
```

```python
df=status(df)
```

<span style="font-family:Calibri; font-size:18pt;">For this step remove the status column</span>

```python
df_final = df.where(df.status ==1).dropna().drop(columns=['status'])
```

```python
df_final
```
